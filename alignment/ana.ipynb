```python
import numpy as np
import matplotlib.pyplot as plt

def plot_cumulative_power(file_path):
    """Plot cumulative power distribution from a .npy file"""
    
    # Load the cumulative power data
    cum_power = np.load(file_path)
    
    # Create figure
    plt.figure(figsize=(10, 6))
    
    # Plot cumulative power
    x = np.arange(1, len(cum_power) + 1)
    plt.plot(x, cum_power, 'b-', linewidth=2)
    
    # Customize plot
    plt.xlabel('Number of Modes', fontsize=12)
    plt.ylabel('Cumulative Power', fontsize=12)
    plt.title('Cumulative Power Distribution', fontsize=14)
    plt.grid(True)
    
    # Optional: Add reference lines at y=0.5 and y=0.9
    plt.axhline(y=0.5, color='r', linestyle='--', alpha=0.5)
    plt.axhline(y=0.9, color='r', linestyle='--', alpha=0.5)
    
    # Show plot
    plt.tight_layout()
    plt.show()
    
    # Print some statistics
    print(f"\nStatistics:")
    print(f"Number of modes needed for 50% power: {np.searchsorted(cum_power, 0.5) + 1}")
    print(f"Number of modes needed for 90% power: {np.searchsorted(cum_power, 0.9) + 1}")

if __name__ == "__main__":
    # You can replace this with your actual file path
    file_path = "/mnt/users/goringn/NNs_vs_Kernels/alignment/msp_cumpower_d30_h400_depth4_n1028_cumpower.npy"
    plot_cumulative_power(file_path)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_cumulative_powers(file_path1, file_path2, 
                          label1="Model 1", label2="Model 2",
                          figsize=(2.0, 7.0),
                          marker_size=10,
                          linewidth=1.5,
                          title_size=8,
                          tick_size=7,
                          legend_size=8):
    """Plot cumulative power distributions with specific formatting"""
    
    # Load the cumulative power data
    cum_power1 = np.load(file_path1)
    cum_power2 = np.load(file_path2)
    
    # Create figure
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    
    # Get x values
    x1 = np.arange(1, len(cum_power1) + 1)
    x2 = np.arange(1, len(cum_power2) + 1)
    
    # Plot lines and scatter points
    ax.plot(x1, cum_power1, color='#4ECDC4', alpha=0.7, linewidth=linewidth)
    ax.plot(x2, cum_power2, color='#FF6B6B', alpha=0.7, linewidth=linewidth)
    
    ax.scatter(x1, cum_power1, color='#4ECDC4', marker='s', s=marker_size, alpha=0.7)
    ax.scatter(x2, cum_power2, color='#FF6B6B', marker='s', s=marker_size, alpha=0.7)
    
    # Show all spines and make them thin
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_linewidth(0.5)
    
    # Set x and y ticks
    x_max = max(len(cum_power1), len(cum_power2))
    x_ticks = [1, x_max]
    y_ticks = [cum_power1[0], 1.0]
    
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)
    
    # Format tick labels
    ax.set_xticklabels([f'{int(val)}' for val in ax.get_xticks()], fontsize=tick_size)
    ax.set_yticklabels([f'{val:.1f}' for val in ax.get_yticks()], fontsize=tick_size)
    
    # Set scales
    ax.set_xscale('log')
    #ax.set_yscale('log')
    
    # Set square aspect ratio
    ax.set_box_aspect(1)
    
    # Make tick marks thinner
    ax.tick_params(width=0.5)
    
    # Add legend
    lines = ax.get_lines()[::2]
    fig.legend(lines, [label1, label2],
              loc='upper center',
              bbox_to_anchor=(0.5, 0.5),
              ncol=2,
              fontsize=legend_size,
              frameon=False)
    
    plt.tight_layout()
    plt.subplots_adjust(top=0.9)
    
    return fig, ax

if __name__ == "__main__":
    # Example usage - replace with your actual file paths
    file_path1 = "/mnt/users/goringn/NNs_vs_Kernels/alignment/cifar10_cumulative_power_1028.npy"
    file_path2 = "/mnt/users/goringn/NNs_vs_Kernels/alignment/msp_cumpower_d30_h400_depth4_n1028_cumpower.npy"
    
    fig, ax = plot_cumulative_powers(
        file_path1=file_path1,
        file_path2=file_path2,
        label1="Depth 2",
        label2="Depth 4"
    )
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_cumulative_powers(file_path1, file_path2, file_path3=None,
                          label1="Model 1", label2="Model 2", label3="Model 3",
                          figsize=(7.0, 7.0),
                          marker_size=10,
                          linewidth=1.5,
                          title_size=8,
                          tick_size=7,
                          legend_size=8):
    """Plot two or three cumulative power distributions with specific formatting"""
    
    # Load the cumulative power data
    cum_power1 = np.load(file_path1)
    cum_power2 = np.load(file_path2)
    
    # Enable LaTeX rendering
    plt.rcParams['text.usetex'] = True
    plt.rcParams['font.family'] = 'serif'
    
    # Create figure
    fig, ax = plt.subplots(1, 1, figsize=figsize)
    
    # Get x values
    x1 = np.arange(1, len(cum_power1) + 1)
    x2 = np.arange(1, len(cum_power2) + 1)
    
    # Plot first two distributions
    ax.plot(x1, cum_power1, color='#4ECDC4', alpha=0.7, linewidth=linewidth)
    ax.plot(x2, cum_power2, color='#FF6B6B', alpha=0.7, linewidth=linewidth)
    
    ax.scatter(x1, cum_power1, color='#4ECDC4', marker='s', s=marker_size, alpha=0.7)
    ax.scatter(x2, cum_power2, color='#FF6B6B', marker='s', s=marker_size, alpha=0.7)
    
    # Add third distribution if provided
    if file_path3 is not None:
        cum_power3 = np.load(file_path3)
        x3 = np.arange(1, len(cum_power3) + 1)
        ax.plot(x3, cum_power3, color='#45B7D1', alpha=0.7, linewidth=linewidth)
        ax.scatter(x3, cum_power3, color='#45B7D1', marker='s', s=marker_size, alpha=0.7)
        labels = [label1, label2, label3]
    else:
        labels = [label1, label2]
    
    # Show all spines and make them thin
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_linewidth(0.5)
    
    # Set x and y ticks
    x_max = max(len(cum_power1), len(cum_power2))
    if file_path3 is not None:
        x_max = max(x_max, len(cum_power3))
    x_ticks = [1, x_max]
    y_ticks = [0.0, 1.0]
    
    ax.set_xticks(x_ticks)
    ax.set_yticks(y_ticks)
    
    # Format tick labels
    ax.set_xticklabels([f'{int(val)}' for val in ax.get_xticks()], fontsize=tick_size)
    ax.set_yticklabels([f'{val:.1f}' for val in ax.get_yticks()], fontsize=tick_size)
    
    # Set x-scale to log
    ax.set_xscale('log')
    
    # Set square aspect ratio
    ax.set_box_aspect(1)
    
    # Make tick marks thinner
    ax.tick_params(width=0.5)
    
    # Add legend
    lines = ax.get_lines()[::2]
    fig.legend(lines, labels,
              loc='upper center',
              bbox_to_anchor=(0.5, 1.1),
              ncol=len(labels),
              fontsize=legend_size,
              frameon=False)
    
    # Add padding for y-axis labels
    plt.tight_layout()
    plt.subplots_adjust(top=0.9, left=0.15)
    
    return fig, ax

if __name__ == "__main__":
    # Example usage - replace with your actual file paths
    file_path1 = "/mnt/users/goringn/NNs_vs_Kernels/alignment/msp_cumpower_d30_h400_depth4_n4096_cumpower.npy"
    file_path2 = "/mnt/users/goringn/NNs_vs_Kernels/alignment/cifar10_cumulative_power_1028.npy"
    file_path3 = "/mnt/users/goringn/NNs_vs_Kernels/alignment/cifar10ffnn_cumulative_power_4096.npy"  # Optional
    
    # Plot with two distributions
    # fig1, ax1 = plot_cumulative_powers(
    #     file_path1=file_path1,
    #     file_path2=file_path2,
    #     label1="Depth 2",
    #     label2="Depth 4"
    # )
    # plt.show()
    
    # Plot with three distributions
    fig2, ax2 = plot_cumulative_powers(
        file_path1=file_path1,
        file_path2=file_path2,
        file_path3=file_path3,
        label1="Depth 2",
        label2="Depth 4",
        label3="Depth 6"
    )
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler

def setup_science_style():
    """Set up consistent science styling"""
    plt.style.use('default')
    mpl.rcParams.update({
        'figure.figsize': (1.3, 2.5),
        'figure.dpi': 100,
        'font.family': 'serif',
        'font.serif': ['cmr10', 'Computer Modern Serif', 'DejaVu Serif'],
        'text.usetex': False,
        'axes.formatter.use_mathtext': True,
        'mathtext.fontset': 'cm',
        'axes.linewidth': 0.5,
        'axes.spines.top': True,
        'axes.spines.right': True,
        'axes.spines.left': True,
        'axes.spines.bottom': True,
        'xtick.direction': 'in',
        'ytick.direction': 'in',
        'xtick.major.width': 0.5,
        'ytick.major.width': 0.5,
        'xtick.minor.width': 0.5,
        'ytick.minor.width': 0.5,
        'xtick.major.size': 3,
        'ytick.major.size': 3,
        'xtick.minor.size': 1.5,
        'ytick.minor.size': 1.5,
        'xtick.top': True,
        'ytick.right': True,
        'grid.linewidth': 0.5,
        'lines.linewidth': 1.0,
        'lines.markersize': 3,
        'legend.frameon': False,
        'legend.borderpad': 0,
        'legend.borderaxespad': 1.0,
        'legend.handlelength': 1.0,
        'legend.handletextpad': 0.5,
    })

def plot_cumulative_powers(file_path1, file_path2, file_path3=None,
                          label1="Depth 2", label2="Depth 4", label3="Depth 6",
                          figsize=(3.3, 2.5),
                          marker_size=3,
                          linewidth=2.0,
                          n_points=None,
                          save_path=None,
                          dpi=300,
                          bbox_inches='tight',
                          pad_inches=0.02):  # Add saving parameters
    """
    Plot cumulative power distributions with science styling
    
    Parameters:
        file_path1: str - Path to first data file
        file_path2: str - Path to second data file
        file_path3: str, optional - Path to third data file
        label1: str - Label for first dataset
        label2: str - Label for second dataset
        label3: str - Label for third dataset
        figsize: tuple - Figure size in inches
        marker_size: float - Size of markers
        linewidth: float - Width of lines
        n_points: int, optional - Number of points to plot (plots all if None)
        save_path: str, optional - Path to save the figure (if None, figure is not saved)
        dpi: int - DPI for saved figure (default: 300)
        bbox_inches: str - Bounding box setting for saved figure (default: 'tight')
        pad_inches: float - Padding for saved figure (default: 0.02)
    """
    
    # Set up science style
    setup_science_style()
    
    # Load the cumulative power data
    cum_power1 = np.load(file_path1)
    cum_power2 = np.load(file_path2)
    
    # Slice data if n_points is specified
    if n_points is not None:
        cum_power1 = cum_power1[:n_points]
        cum_power2 = cum_power2[:n_points]
    
    # Create figure
    fig, ax = plt.subplots(figsize=figsize)
    
    # Get x values
    x1 = np.arange(1, len(cum_power1) + 1)
    x2 = np.arange(1, len(cum_power2) + 1)
    
    # Plot first two distributions
    ax.plot(x1, cum_power1, color='#96FFC0', alpha=0.7, linewidth=linewidth,
            marker='', markersize=marker_size, label=label1)
    ax.plot(x2, cum_power2, color='#FF0000', alpha=0.7, linewidth=linewidth,
            marker='', markersize=marker_size, label=label2)
    
    # Add third distribution if provided
    if file_path3 is not None:
        cum_power3 = np.load(file_path3)
        if n_points is not None:
            cum_power3 = cum_power3[:n_points]
        x3 = np.arange(1, len(cum_power3) + 1)
        ax.plot(x3, cum_power3, color='#7A00FF', alpha=0.7, linewidth=linewidth,
                marker='', markersize=marker_size, label=label3)
    
    # Set x and y ticks
    x_max = max(len(cum_power1), len(cum_power2))
    if file_path3 is not None:
        x_max = max(x_max, len(cum_power3))
    
    # Configure axes
    ax.set_xscale('log')
    #ax.set_yscale('log')
    ax.minorticks_on()
    
    # Set axis labels
    ax.set_xlabel(r'$\rho$', labelpad=2)
    ax.set_ylabel(r'$C(\rho)$', labelpad=2)
    
    # Add legend inside plot
    ax.legend(loc='lower right', fontsize=7, frameon=False,
             handlelength=1.0, handletextpad=0.5)
    
    # Set square aspect ratio and adjust layout
    ax.set_box_aspect(0.35)
    plt.tight_layout()
    
    # Save figure if save_path is provided
    if save_path is not None:
        plt.savefig(save_path, dpi=dpi, bbox_inches=bbox_inches, pad_inches=pad_inches)
    
    return fig, ax

if __name__ == "__main__":
    # Example usage - replace with your actual file paths
    file_path1 = "/mnt/users/goringn/NNs_vs_Kernels/alignment/msp_cumpower_d30_h400_depth4_n4096_cumpower.npy"
    file_path2 = "/mnt/users/goringn/NNs_vs_Kernels/alignment/cifar10_cumulative_power_4096.npy"
    file_path3 = "/mnt/users/goringn/NNs_vs_Kernels/alignment/cifar10ffnn_cumulative_power_4096.npy"
    
    # Plot only first 100 values and save the figure
    fig, ax = plot_cumulative_powers(
        file_path1=file_path1,
        file_path2=file_path2,
        file_path3=file_path3,
        label1="FFNN on MSF ",
        label2="CNN on CIFAR10",
        label3="FFNN on CIFAR10",
        n_points=500,  # Specify number of points to plot
        save_path='cumulative_powers.pdf'  # Save figure as PDF
    )
    plt.show()
```
