```python
import matplotlib.pyplot as plt
import pickle
import os

def plot_dict_of_lists(data_dict, x_values=None, xlabel='X-axis', ylabel='Y-axis',
                      title='Plot', log_x=False, log_y=False):
    """
    Plots a dictionary of lists with a custom x-axis and optional logarithmic scaling.
    """
    plt.figure(figsize=(10, 6))

    if x_values is None:
        x_values = range(len(next(iter(data_dict.values()))))

    for label, values in data_dict.items():
        # Convert ratio to percentage for label
        ratio_pct = float(label) * 100
        plt.plot(x_values, values, marker='o', label=f"Pruned {ratio_pct}%")

    if log_x:
        plt.xscale('log')
    if log_y:
        plt.yscale('log')

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.legend(loc='best')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()

def load_and_plot_pruning_results(results_file):
    """
    Load pruning results from file and create plots.
    """
    # Load results
    print(f"Loading results from {results_file}")
    with open(results_file, 'rb') as f:
        results = pickle.load(f)
    
    # Get n_train sizes and pruning ratios from results
    n_train_sizes = sorted(list(results['original']['test'].keys()))
    pruning_ratios = [key.split('_')[1] for key in results.keys() if key.startswith('pruned')]
    
    print(f"Found {len(n_train_sizes)} training sizes: {n_train_sizes}")
    print(f"Found {len(pruning_ratios)} pruning ratios: {pruning_ratios}")
    
    # Calculate percentage differences for test set
    percent_diff_test = {}
    for ratio in pruning_ratios:
        ratio_key = f'pruned_{ratio}'
        percent_diff_test[ratio] = [
            (results[ratio_key]['test'][n] - results['original']['test'][n]) / results['original']['test'][n] * 100 
            for n in n_train_sizes
        ]
    
    # Plot test differences
    plot_dict_of_lists(
        percent_diff_test,
        x_values=n_train_sizes,
        xlabel='Number of training samples',
        ylabel='Percentage difference in MSE loss',
        title='Test Loss Difference vs Original Model',
        log_x=True,
        log_y=True
    )
    
    # Calculate and plot training differences
    percent_diff_train = {}
    for ratio in pruning_ratios:
        ratio_key = f'pruned_{ratio}'
        percent_diff_train[ratio] = [
            (results[ratio_key]['train'][n] - results['original']['train'][n]) / results['original']['train'][n] * 100 
            for n in n_train_sizes
        ]
    
    plot_dict_of_lists(
        percent_diff_train,
        x_values=n_train_sizes,
        xlabel='Number of training samples',
        ylabel='Percentage difference in MSE loss',
        title='Training Loss Difference vs Original Model',
        log_x=True,
        log_y=True
    )

if __name__ == "__main__":
    # Specify your results file path here
    results_file = "/mnt/users/goringn/NNs_vs_Kernels/stair_function/results/msp_anthro_false_mup_lr0005_gamma_001_modelsaved/pruning_results_20250127_205324.pkl"
    load_and_plot_pruning_results(results_file)
```
