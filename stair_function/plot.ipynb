```python
import json
import matplotlib.pyplot as plt
import numpy as np
import os

def load_and_plot_combined(kernel_path, ensemble_path1, ensemble_path2, labels=None):
    # Read the JSON files
    with open(kernel_path, 'r') as f:
        kernel_data = json.load(f)
    with open(ensemble_path1, 'r') as f:
        ensemble_data1 = json.load(f)
    with open(ensemble_path2, 'r') as f:
        ensemble_data2 = json.load(f)
    
    # Extract kernel method data
    n_train_kernel = [d['n_train'] for d in kernel_data]
    sgd_errors = [d['sgd_test_error'] for d in kernel_data]
    nngp_errors = [d['nngp_matched_error'] for d in kernel_data]
    ntk_errors = [d['ntk_matched_error'] for d in kernel_data]
    
    # Extract ensemble data
    n_train1 = [d['n_train'] for d in ensemble_data1]
    ensemble_errors1 = [d['ensemble_test_error'] for d in ensemble_data1]
    
    n_train2 = [d['n_train'] for d in ensemble_data2]
    ensemble_errors2 = [d['ensemble_test_error'] for d in ensemble_data2]
    
    # Create the plot
    plt.figure(figsize=(5, 3))
    
    # Plot all results
    plt.plot(n_train_kernel, sgd_errors, 'o-', color='#7400FF',
             linewidth=3, markersize=2, label='SGD')
    plt.plot(n_train_kernel, nngp_errors, 'o-', color='#89FFB9',
             linewidth=3, markersize=2, label='NNGP')
    plt.plot(n_train_kernel, ntk_errors, 'o-', color='#FF89E7',
             linewidth=3, markersize=2, label='NTK')
    
    ensemble1_label = labels[0] if labels else 'Ensemble 1'
    ensemble2_label = labels[1] if labels else 'Ensemble 2'
    
    plt.plot(n_train1, ensemble_errors1, 'o-', linewidth=3, markersize=2, label=ensemble1_label)
    plt.plot(n_train2, ensemble_errors2, 'o-', linewidth=3, markersize=2, label=ensemble2_label)
    
    # Set both scales to log
    plt.xscale('log')
    plt.yscale('log')
    
    # Labels and styling
    plt.xlabel('Training Set Size')
    plt.ylabel('Test Error')
    plt.tick_params(axis='both', labelcolor='grey', size=0)
    plt.legend()
    plt.title("Test Error Comparison")
    
    # Remove grid
    plt.grid(False)
    
    # Show plot
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    kernel_path = "/mnt/users/goringn/NNs_vs_Kernels/stair_function/results/results_20241130_042001.json"
    ensemble_path1 = "/mnt/users/goringn/NNs_vs_Kernels/stair_function/results/long_epoch/results_20241202_115751.json"
    ensemble_path2 = "/mnt/users/goringn/NNs_vs_Kernels/stair_function/results/long_epoch/results_20241202_160316.json"
    
    labels = ["T=0.0005", "T=0.005"]
    
    load_and_plot_combined(kernel_path, ensemble_path1, ensemble_path2, labels)
```
